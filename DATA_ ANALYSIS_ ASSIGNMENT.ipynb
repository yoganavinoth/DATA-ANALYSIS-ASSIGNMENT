# 8)Test the similarity between the degree_t(Sci&Tech) and specialization(Mkt&HR) with respect to salary at significant level of 5%(Make decision using Hypothesis testing)

Independent Sample-Unpaired T Test. Different group(degree_t,specialisation)but same condition(salary)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
degree=dataset[dataset["degree_t"]=="Sci&Tech"]["salary"]
specialisation=dataset[dataset["specialisation"]=="Mkt&HR"]["salary"]

ttest_ind(degree,specialisation)
```

Hypothesis testing. Statement:Test the similarity between the degree_t(Sci&Tech)and specialisation(Mkt&HR) with respect to salary. Alternative hypothesis -there is similarity between degree_t of sci&Tech and specialisation of mkt&hr

solution: P value is less than 5%.so it accept Alternative Hypothesis and reject Null hypothesis.

# 9)Convert the normal distribution to standardnormal distribution for salary columns

```python
#!pip install seaborn
```

```python
#Standard Normalisation
def stdNBgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    #formula calculation
    z_score=[((j-mean)/std) for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
stdNBgraph(dataset["salary"])
```

# 10)What is the probability Density Function of the salary range from 700000 to 900000 ?

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'purple'},color='orange')
    pyplot.axvline(startrange,color='green')
    pyplot.axvline(endrange,color='green')
    sample=dataset
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f , Standard Deviation=%.3f' % (sample_mean, sample_std))
    dist=norm(sample_mean,sample_std)
    values=[value for value  in range (startrange,endrange)]
    probabilities=[dist.pdf(values) for values in values] 
    prob=sum(probabilities)
    print ("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```
